```python
import numpy as np
```

```python
image = np.random.randn(32, 32, 3)
kernel = np.random.randn(5, 5, 3)
```

```python
out_h, out_w = image.shape[0] - kernel.shape[0] + 1, image.shape[0] - kernel.shape[0] + 1
```

```python
out = np.zeros((out_h, out_w))
```

```python
for i in range(out_h):
    for j in range(out_w):
        patch = image[i:i+5, j:j+5, :]
        out[i, j] = np.sum(patch * kernel)

print(out)
print(out.shape)
```
