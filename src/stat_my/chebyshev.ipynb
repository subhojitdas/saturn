```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
%matplotlib inline
```

```python
np.random.seed(18)
data = np.random.exponential(scale=10, size=10000)
data
```

```python
plt.plot(data)
```

```python
mean = np.mean(data)
std = np.std(data)

k_values = [1,2,3]
bounds = [(mean - k * std, mean + k * std) for k in k_values]

plt.figure(figsize=(12, 6))
sb.histplot(data, bins=100, kde=True, color="grey", stat="density")

plt.axvline(mean, color='blue', linestyle='--', label=f'Mean = {mean:.2f}')
for k, (low, high) in zip(k_values, bounds):
    plt.axvline(low, color='red', linestyle='--', label=f'{k}Ïƒ bound')
    plt.axvline(high, color='red', linestyle='--')

plt.title("Chebyshev's Inequality Demonstration on Skewed Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
