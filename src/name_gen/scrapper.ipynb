```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.babynamesofindia.com/hindu-baby-boy-names-a.html'  # Example page

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

names = []
for row in soup.select('table tr')[1:]:  # Skip header
    cols = row.find_all('td')
    if len(cols) >= 3:
        name = cols[0].get_text(strip=True)
        gender = cols[1].get_text(strip=True)
        meaning = cols[2].get_text(strip=True)
        names.append([name, gender, meaning, 'Hindu', 'Boy'])

# Save to CSV
with open('indian_names.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Name', 'Gender', 'Meaning', 'Religion', 'Category'])
    writer.writerows(names)
```
